```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
balloon_length = np.random.normal(30, 5, 200)
balloon_width = np.random.normal(40, 5, 200)
balloon_weight = np.random.normal(0.005, 0.005, 200)
```

```python
melon_length = np.random.normal(18, 4, 200)
melon_width = np.random.normal(15, 4, 200)
melon_weight = np.random.normal(2.7, 0.2, 200)
```

```python
fig = plt.figure(figsize = (6,6))
ax = plt.axes(projection='3d')

ax.scatter(xs=balloon_width, ys=balloon_length, zs=balloon_weight, color='red', marker="^", label='balloon')
ax.scatter(xs=melon_width, ys=melon_length, zs=melon_weight, color='green', marker="o", label='melon')

ax.grid()
ax.set_zlabel('weight (kg)',  labelpad=-1)

plt.title('Balloon and melon classification')
plt.xlabel("width (cm)")
plt.ylabel("length (cm)")

plt.legend()
plt.savefig("balloon_melon.jpg")
plt.show()
```
