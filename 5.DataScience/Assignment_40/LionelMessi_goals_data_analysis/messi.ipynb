```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("Lionel Messi Goals.csv")
data.head()
```

```python
df = data["Goal Type"].value_counts().reset_index()
df
```

```python
plt.pie(df["count"], labels=df["Goal Type"], autopct='%1.1f%%', shadow=True, startangle=90)
plt.title("Messi's goals types")
plt.show()
```

```python
df = data["Goal Method"].value_counts().reset_index()
df
```

```python
plt.bar(df["Goal Method"], df["count"])
plt.title("Messi's goals methods")
plt.show()
```

```python
data["Minute"] = data["Minute"].apply(eval)
data["Minute"] = data["Minute"].astype(int)
df = data["Minute"].value_counts().reset_index().sort_values(by=["Minute"], ascending=True)
df
```

```python
plt.figure(figsize=(24, 8))
plt.plot(df["Minute"], df["count"])
plt.title("Messi's goals minutes")
plt.show()
```

```python
data1 = data
data1['Date'] = data1['Date'].to_numpy(str)

date_list = []
for date in data1['Date'] :
    date_list.append( date[-4:])
    
data1['Date'] = date_list

data1.head()
```

```python
df = data1.value_counts(['Date']).reset_index()
df = df.sort_values(by= 'Date' , ascending=True).reset_index()
df
```

```python
plt.figure(figsize=(24, 8))
plt.bar(df["Date"], df["count"])
plt.xlabel("year")
plt.ylabel("number of goals")
plt.title("Number of Messi's goals per year")
plt.savefig("output/1.jpg")
plt.show()
```

```python
data1.rename(columns = {'Goal Type':'GoalType'}, inplace = True)

df = data1.query("GoalType == 'Penalty'")['Goal Method'].value_counts().reset_index()
df.loc[1]=['Right foot',0]
df
```

```python
plt.bar(df["Goal Method"], df["count"])
plt.xlabel("right or left foot")
plt.ylabel("number of goals")
plt.title("Messi's penalty goals")
plt.savefig("output/2.jpg")
plt.show()
```

```python
data = pd.read_csv("Lionel Messi Goals.csv")
data.head()
```

```python
data["Score Home Team - When Messi Scored"] = data["Score Home Team - When Messi Scored"].astype(int)
data["Score Away Team - When Messi Scored"] = data["Score Away Team - When Messi Scored"].astype(int)
```

```python
data['Is_Home_Game'] = data['Home team'].apply(lambda x: 'FC Barcelona' in x or 'Argentina' in x)
```

```python
data['Is_Draw_Goal'] = (data['Score Home Team - When Messi Scored'] == data['Score Away Team - When Messi Scored'])
data.head()
```

```python
draw_goals_count = data['Is_Draw_Goal'].sum()
total_goals_count = len(data)
draw_percentage = (draw_goals_count / total_goals_count) * 100

print(f"Percentage of Messi's goals that were equalizer goals: {draw_percentage:.2f}%")
```
