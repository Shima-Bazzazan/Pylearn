```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("input/COVID-19_global_April.csv")

data.head(10)
```

```python
data['Date'] = pd.to_datetime(data['Date'])

data['YearMonth'] = data['Date'].map(lambda date: 100*date.year + date.month)

data['April 2020 new cases'] = data['New cases']
data1 = data.groupby(['Country/Region'])['April 2020 new cases'].sum().reset_index()
data1 = data1.sort_values(by='April 2020 new cases', ascending=False)
data1.head(8)
```
