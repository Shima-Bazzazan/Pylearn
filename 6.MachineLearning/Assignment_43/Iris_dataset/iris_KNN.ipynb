```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn import metrics
import seaborn as sns
from knn import KNN
```

```python
iris = load_iris()
df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
df.head()
```

```python
df['targets'] = pd.Series(iris.target)
df['Iris_name'] = df['targets'].apply(lambda x: 'setosa' if x == 0 else 
                                      ('versicolor' if x == 1 else 'virginica'))
df
```

```python
df["Iris_name"] = df["Iris_name"].replace(["setosa", "versicolor", "virginica"], [0, 1, 2]) 
```

```python
cmap = ["red", "blue", "green"]
_, ax_1 = plt.subplots()
scatter = ax_1.scatter(iris.data[:, 0], iris.data[:, 1], c=iris.target, cmap=ListedColormap(cmap))
ax_1.set(xlabel=iris.feature_names[0], ylabel=iris.feature_names[1])
_ = ax_1.legend(scatter.legend_elements()[0], iris.target_names, loc="lower right", title="Classes")
```

```python
cmap = ["red", "blue", "green"]
_, ax_2 = plt.subplots()
scatter = ax_2.scatter(iris.data[:, 2], iris.data[:, 3], c=iris.target, cmap=ListedColormap(cmap))
ax_2.set(xlabel=iris.feature_names[2], ylabel=iris.feature_names[3])
_ = ax_2.legend(scatter.legend_elements()[0], iris.target_names, loc="lower right", title="Classes")
```

```python
X = np.array([df["sepal length (cm)"], df["sepal width (cm)"], df["petal length (cm)"], 
              df["petal width (cm)"]]).T
y = np.array(df["Iris_name"], dtype=int)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=None)
```

```python
knn = KNN(k=7)
knn.fit(X_train, y_train)
```

```python
new_iris = np.array([4.5, 3.3 , 1.4, 0.1])
new_iris_2 = np.array([5.4, 3.9, 1.7, 1.4])
new_iris_3 = np.array([6.1, 3.0 , 5.0, 2.1])

new_irises = [new_iris, new_iris_2, new_iris_3]

outputs = knn.predict(new_irises)
```

```python
for output in outputs:
    if output == 0:
        print("setosa")
    elif output == 1:
        print("versicolor")
    else:
        print("virginica")
```

```python
knn.evaluate(X_test, y_test)
```

```python
y_pred = knn.predict(X_test)
```

```python
confusion_matrix = metrics.confusion_matrix(y_test, y_pred)
cm_display = metrics.ConfusionMatrixDisplay(confusion_matrix = confusion_matrix, display_labels = ['setosa', 'versicolor', 'virginica'])
fig, ax = plt.subplots(1, 1, figsize=(7, 7))
ax = sns.heatmap(confusion_matrix, annot=True, square=True , xticklabels=iris.target_names  ,  yticklabels=iris.target_names  ,  fmt='g', cmap="flare" , annot_kws={"size": 15})
ax.set_xlabel('Predicted')
ax.set_ylabel('Actual')
ax.set_title('Confusion Matrix for the Iris Dataset')
plt.show()
```
