```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.model_selection import train_test_split
from lls import LLS
from scipy.stats import linregress
```

Generating continuous random dataset for the problem:

```python
X, y, coef = datasets.make_regression(n_samples=200,n_features=1,noise=30,coef=True , random_state=20) 

X = np.interp(X, (X.min(), X.max()), (0.5, 8))
y = np.interp(y, (y.min(), y.max()), (0, 20))
```

```python
plt.scatter(X, y, alpha= 0.5)
plt.legend()
plt.xlabel("Study Duration")
plt.ylabel("Student's Grade")
plt.title('Impact of Study Hours on Student Grades')
```

Splitting dataset to train and test datasets:

```python

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=None)
```

Using my object oriented linear least squares (LLS) class:

```python
lls = LLS()
slp = lls.slope(X_train, y_train)
slp
```

```python
plt.scatter(X_train, y_train, label='Data', alpha= 0.5)
plt.plot(X_train, slp * X_train, color='red', label='Linear Regression')
plt.xlabel("Study Duration")
plt.ylabel("Student's Grade")
plt.title('Impact of Study Hours on Student Grades')
plt.legend()
plt.show()
```

```python
slop, intercept, rvalue, pvalue, stderr =linregress(X_train.T, y_train)
```
